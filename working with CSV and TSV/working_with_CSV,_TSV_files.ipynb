# **Import data from URL**


```python
########################### Older version of importing files from an URL ##########################################

import requests
from io import StringIO

# The URL of the CSV file
url = "https://raw.githubusercontent.com/cs109/2014_data/master/countries.csv"

# Send an HTTP request to get the content of the CSV file
headers = {"User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10.14; rv:66.0) Gecko/20100101 Firefox/66.0"}
req = requests.get(url, headers=headers)

# Use StringIO to create a file-like object from the CSV content
data = StringIO(req.text)

pd.read_csv(data)
```

```python
##################### Squeeze Parameter ###########################
pd.read_csv('aug_train.csv',usecols=['gender'],squeeze=True)
```

```python
###########SkipRows###############


# Replace 'your_file.csv' with the actual path to your CSV file
file_path = 'aug_train.csv'

# Define a custom function to skip rows that are multiples of 5
def skip_rows_func(index):
    return index % 5 == 0

# Read the CSV file into a Pandas DataFrame with custom skiprows function
df = pd.read_csv(file_path, skiprows=skip_rows_func)

# Display the first few rows of the dataset
print(df.head())
```

```python
##################### Encoding Parameter ###################
pd.read_csv('zomato.csv',encoding='latin-1')
```

```python
##################### Handling Dates ###################
pd.read_csv('IPL Matches 2008-2020.csv',parse_dates=['date']).info()
```

```python
##################### Converters ###################
def rename(name):
    if name == "Royal Challengers Bangalore":
        return "RCB"
    else:
        return name
rename("Royal Challengers Bangalore")
```

```python
pd.read_csv('IPL Matches 2008-2020.csv',converters={'team1':rename})
```

```python
############################## Load huge dataset in chunks #########################################

dfs = pd.read_csv('aug_train.csv',chunksize=5000)
for chunks in dfs:
    print(chunks.shape)
```
